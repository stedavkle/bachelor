```python
import os
import SEM_API_CUSTOM
import nanocontrol
import random
import itertools
```

```python
# Initialize SEM
sem = SEM_API_CUSTOM.SEM_API_CUSTOM()
sem.openConnection()
(magnification, rotation, detector, scanrate, workingdistance) = sem.getInitialParameters()
# initialize nanocontrol
controller = nanocontrol.controller()
```

```python
# Initialize Parameters
dataset_path = r"C:\Users\david\Desktop\automated"
magnifications = [50000, 25000, 10000, 2000]
retract_factors = [1, 1, 5, 8]
wds = [0.0090153, 0.009057] # tips, stage
scanrates = ['2', '6']
detectors = ['SE2', 'InLens']
```

```python
for i in range(0, len(magnifications)):
    controller.assignPattern([])
    for j in range(0, 5): # 5 scenes in each magnification setting
        print('mask: %s' % (i*5+j))
        # mask
        sem.grabImageWithParameters(os.path.join(dataset_path, 'masks', '%s%s.tif' % (i,j)), magnifications[i], rotation, 'SE2', '10', wds[0])
        # images
        #input('mask ok?')
        for k, (wd, sr, dt) in enumerate(random.sample(list(itertools.product(wds, scanrates, detectors)), 5)):
            print('image: %s' % k, wd, sr, dt)
            #input('image ok?')        
            sem.grabImageWithParameters(os.path.join(dataset_path, 'images', '%s%s%s.tif' % (i, j, k)), magnifications[i], rotation, dt, sr, wd)
            #controller.retractStep(retract_factors[i])
            controller.moveStage()
        controller.retractStep(retract_factors[i])
```

```python
sem.restoreInitialParameters()
sem.closeConnection()
controller.closeAll()
```
